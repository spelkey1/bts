```python
import datetime
import os

import pandas as pd
import numpy as np
pd.options.display.max_columns = 200
```

```python
df = pd.read_csv('final_dataset_complete.csv', index_col=0)
df['date'] = pd.to_datetime(df.date)
df.sort_values('date', inplace=True)
```

```python
# get rolling hitter vs pitcher stats for single season
gb = ['results_|_Season', 'team_fg','results_|_Name','opp_team_fg', 'opp_pitcher']
cols = ['results_|_PA', 'results_|_H', ]
df_gb = df[gb + cols].groupby(gb).cumsum()
cols = ['pa_cum', 'hit_cum']
df_gb.columns = cols
df = pd.concat([df, pd.concat([df, df_gb], axis = 1)[gb + cols].groupby(gb).shift()], axis = 1)
df['avg_cum'] = df.hit_cum / df.pa_cum
df.drop('hit_cum', axis = 1, inplace = True)
```

```python
# get rolling hitter vs pitcher stats for all three seasons
# due to people changing teams,and having no unique id, this will not be the best measure
gb = ['results_|_Name', 'opp_pitcher']
cols = ['results_|_PA', 'results_|_H', ]
df_gb = df[gb + cols].groupby(gb).cumsum()
cols = ['pa_cum_p3', 'hit_cum_p3']
df_gb.columns = cols
df = pd.concat([df, pd.concat([df, df_gb], axis = 1)[gb + cols].groupby(gb).shift()], axis = 1)
df['avg_cum_p3'] = df.hit_cum_p3 / df.pa_cum_p3
df.drop('hit_cum_p3', axis = 1, inplace = True)
```

```python
df.fillna(-1, inplace = True)
df.to_csv('final_dataset_complete2.csv')
```
